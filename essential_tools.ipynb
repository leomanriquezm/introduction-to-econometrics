##Promedio y desviación estándar
Recordemos que $\bar{X}=\frac{\sum_{i=1}^{n}X_{i}}{n}$ y $\sigma_{X}^{2}=\frac{\sum_{i}^{n}(X_{i}-\bar{X})^2}{n}$.

```python
a=np.arange(1,10,1)
#valor mínimo
a
```

```python
#valor máximo
a.max()
```

```python
#media
a.mean()
```

```python
#cálculo manual de media
a.sum()/len(a)
```

```python
#mediana
np.median(a)
```

```python
#varianza
a.var()
```

```python
#cálculo manual de varianza
sum((a-np.mean(a))**2)/len(a)
```

```python
#desviación estándar
a.std()
```

```python
#cálculo manual de desviación estándar
(sum((a-np.mean(a))**2)/len(a))**0.5
```

##Fusionando dataframes
El proceso de recolección de información de un fenómeno no siempre se realiza en un sitio único. Por lo tanto, es necesario saber fusionar bases de datos obtenidas de distintas fuentes. Es importante mencionar que deberá existir un identificador $id$ para hacer match entre los diferentes conjuntos de información.

```python
df1=pd.DataFrame({"id":["Tom","Pedro","Andrea"],"score":[7,5,3]})
df1
```

```python
df2=pd.DataFrame({"id":["Fernanda","Pedro","Andrea"],"age":[15,14,18]})
df2
```

```python
df_final = pd.merge(df1,df2,on="id",how="left")
df_final
```

```python
df_final = pd.merge(df1,df2,on="id",how="right")
df_final
```

```python
df_final = pd.merge(df1,df2,on="id",how="outer")
df_final
```

```python
df_final = pd.merge(df1,df2,right_index=True,left_index=True)
df_final
```

##Concatenando
Permite unir vectores, ya sea verticalmente o adjuntar columnas.

```python
df1=pd.DataFrame(np.random.rand(7),columns=["first"])
df1
```

```python
df2=pd.DataFrame(np.random.rand(7),index=[1,2,3,4,5,6,7],columns=["first"])
df2
```

```python
#en este ejemplo se concatenan verticalmente, notar que df1 y df2 se "unen" en un sólo vector
df_final=pd.concat([df1,df2],axis=0)
df_final
```

```python
#en este ejemplo se concatenan df1 y df2 pero el resultado tienen 2 columnas. La diferencia entre el concatenamiento anterior es que en este ejemplo se "pegan" las columnas unas con otras, según el orden que se indique
df_final=pd.concat([df1,df2],axis=1)
df_final
```
