El modelo de regresión lineal *simple* puede ser expresado como:
\begin{equation}
Y_{i}=\beta_{0}+\beta_{1}\cdot X_{i}+\mu_{i},
\tag{1}
\end{equation}

donde $Y_{i}$ es la variable a explicar, $X_{i}$ es la variable explicativa y $\mu_{i}$ es término del error. Los parámetros $\beta_{0},\beta_{1}$ son parámetros **desconocidos** que estamos interesados en estimar a través de una muestra.

Si por un momento no consideramos el término del error ($\mu_{i}$), entonces el modelo de regresión lineal *simple* puede ser expresado como la ecuación de la recta:

\begin{equation}
Y=10+1.5\cdot X,
\tag{2}
\end{equation}

donde estamos asumiendo que $\beta_{0}=10$ y $\beta_{1}=1.5$.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(1, 10, 15)
Y = 10 + 1.5*X
fig, ax = plt.subplots(figsize = (7, 7))
ax.plot(X, Y)
ax.scatter(X, Y, c ='r')
ax.grid()
ax.set_title('$Y=10+1.5x$')
ax.set_xlim(0, 11)
ax.set_ylim(0, 30)
plt.show()
```

## Mínimos cuadrados ordinarios

El método de **Mínimos Cuadrados Ordinarios** (MCO)...



```python
beta1, beta2 = 10, 1.5
def gen_linreg_data(beta1, beta2, samp_size, disturb_scale):

    X = np.linspace(1, 10, samp_size)
    u = disturb_scale * np.random.randn(samp_size)
    Y = beta1 + beta2*X + u
    Y_hat = beta1 + beta2*X
    return X, Y, Y_hat

def plot_lin_reg(X, Y, Y_hat):
    fig, ax = plt.subplots(figsize = (7, 7))

    for i in range(len(Y)):
        dot_fit_values = [X[i], X[i]]
        dot_org_values = [Y[i], Y_hat[i]]
        ax.plot(dot_fit_values, dot_org_values, linestyle = '--', color = 'red', label = 'residual')

    ax.plot(X, Y_hat)
    ax.scatter(X, Y_hat, c = 'k')
    ax.scatter(X, Y, c ='r')
    ax.grid()
    ax.set_title('$\hat Y ={}+{}X$'.format(beta1, beta2))
    plt.show()

if __name__ == '__main__':
    X, Y, Y_hat = gen_linreg_data(beta1=beta1, beta2=beta2, samp_size=15, disturb_scale=5)
    plot_lin_reg(X, Y, Y_hat)
```

\begin{equation}
\epsilon_{i}=Y_{i}-\hat{Y_{i}}=Y_{i}-\hat{\beta_{0}}-\hat{\beta_{1}}\cdot X_{i}.
\end{equation}

El *core* del $MCO$ consiste en encontrar los estimadores para $\beta_{0}$ y $\beta_{1}$ tal que:

\begin{equation}
\min_{\hat{\beta_{0}},\hat{\beta_{1}}}SRC=\sum_{i}^{n}\epsilon_{i}^{2},
\tag{3}
\end{equation}

donde $SRC$ es la Suma de Residuos Cuadrados
